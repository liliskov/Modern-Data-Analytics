# Function to return Dataframe from a HDF file 

```python
def hdf_to_dataframe(hdf_file_path,year,csv):
    try:
        import pandas as pd 
        import h5py
        import numpy as np
        import os
        import tables
    except:
        print("Missing Library")
    year = str(year)
    file_n = hdf_file_path.split('.')
    Name = file_n[1]
    A = h5py.File(hdf_file_path, 'r')
    list_1 = list(A.keys())
    list_df = pd.DataFrame()
    list_df2 = pd.DataFrame()
    list_df["keys"] = list_1
    lat = A['lat']   
    lon = A['lon'] 
    data_pd = pd.DataFrame()   
    list_df2["keys"] = list_df["keys"].loc[0:37]
    count = 0
    for keys in list_df2["keys"]:
          if keys == 'ACond_tavg':
              data_arr = A[keys][0,:,:]
              data_pd = pd.DataFrame(data = data_arr)
              data_pd.columns = lon
              data_pd.index = lat
              col_n = year + keys
              data_pd = pd.melt(data_pd, ignore_index = False, var_name='lon', value_name= col_n)
              print(keys)
          else:
              data_arr = A[keys][0,:,:]
              data = pd.DataFrame(data = data_arr)
              data.index = lat
              col_n = year + keys
              data = pd.melt(data, ignore_index = False, value_name= col_n)
              data_pd[col_n] = data.loc[:,col_n]
    if csv == 1:
          data_pd['lat'] = data_pd.index
          yearM2 = year+'/'+Name + '.csv'
          data_pd.to_csv(yearM2)
    return data_pd


output = hdf_to_dataframe('GLDAS_CLSM10_M.A200001.021.hdf')
```

```python
def month_csv_to_year_csv(year):
    from os import listdir
    print("Current working directory: {0}".format(os.getcwd()))
    path = 'D:\\MDA\\GLDAS_'+str(year)
    try:
        os.chdir(path)
        print("Current working directory: {0}".format(os.getcwd()))
    except FileNotFoundError:
        print("Directory: {0} does not exist".format(path))
    except NotADirectoryError:
        print("{0} is not a directory".format(path))
    except PermissionError:
        print("You do not have permissions to change to {0}".format(path))
    filepaths = [f for f in listdir("./") if f.endswith('.csv')]
    df = pd.concat(map(pd.read_csv, filepaths)).groupby(['lon','lat']).mean()
    filename = "GLDAS_year_mean_"+str(year)+".csv"
    df.to_csv(filename)
    
for j in range(2000,2014):
    month_csv_to_year_csv(j)
```
